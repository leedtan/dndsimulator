```python
from dice import Dice

d6 = Dice(6)
d8 = Dice(8)

def three_d6():
    return sum([d6.roll() for i in range(3)])
    
def four_d6_drop_low():
    rolls = [d6.roll() for i in range(4)]
    rolls.remove(min(rolls))
    return sum(rolls)
    
def two_d6_add6():
    rolls = d6.roll(2)
    return rolls +6
    
def d8_add10():
    rolls = d8.roll()
    return rolls + 10

def ability_gen_three_d6():
    return [three_d6() for i in range(6)]

def ability_gen_four_d6_drop_low():
    return [four_d6_drop_low() for i in range(6)]

def ability_gen_two_d6_add6():
    return [two_d6_add6() for i in range(6)]

def ability_gen_d8_add10():
    return [d8_add10() for i in range(6)]
```

```python
ability_gen_four_d6_drop_low()
```
