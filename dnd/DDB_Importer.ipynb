```python
import pandas as pd
import requests
```

```python
def get_pc_id(link):
    last_forward = [i for i, char in enumerate(link) if char == '/']
    start_id = last_forward[len(last_forward)-1]
    return link[start_id + 1:]

def get_ddb_pc_json(link):
    json_link_start = 'https://character-service.dndbeyond.com/character/v3/character/'
    pc_id = get_pc_id(link)
    r = requests.get(json_link_start + pc_id)
    return r.text
    
```

```python
test_pc_link = 'https://www.dndbeyond.com/profile/jsanc525/characters/44183476'
test_pc_json = get_ddb_pc_json(test_pc_link)
```
